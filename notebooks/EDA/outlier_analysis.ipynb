```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns   


df = pd.read_csv('weekly_spread.csv')
```

```python
plt.figure(figsize=(15,6))
sns.histplot(df[(df['weekly_avg_spread_bps'] <= 300)],bins=300)
```

```python
plt.figure(figsize=(15,6))
sns.histplot(df['weekly_avg_spread_bps'], bins=300)
```

```python
df = df[(df['weekly_avg_spread_bps'] <= 300) & (df['weekly_avg_spread_bps'] > 0)]




plt.figure(figsize=(15,6))
sns.histplot(df['weekly_avg_spread_bps'], bins=300)
```

```python
plt.figure(figsize=(15,6))

sns.histplot(np.log(df['weekly_avg_spread_bps']), bins=500)
```

```python
df.describe()
```

```python
weekly_spread_clean = df[
    (df['weekly_avg_spread_bps'] > 0) &
    (df['weekly_avg_spread_bps'] <= 300) &
    (df['n_pairs'] >= 3)    
]

sns.histplot(weekly_spread_clean['weekly_avg_spread_bps'], bins=300)
```

```python
plt.figure(figsize=(15,6))

sns.histplot(np.log(weekly_spread_clean['weekly_avg_spread_bps']), bins=500)
```
